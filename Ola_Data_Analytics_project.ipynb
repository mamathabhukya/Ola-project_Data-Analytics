```python
import pandas as pd
import numpy as np

```

```python
ola_data = pd.read_csv("/content/test.csv")
```

```python
ola_data.head()
```

```python
ola_data.info()
```

```python
ola_data['pickup_time'] = pd.to_datetime(ola_data['pickup_time'])
ola_data['drop_time'] = pd.to_datetime(ola_data['drop_time'])
```

```python
ola_data.info()
```

```python
ola_data = ola_data.drop_duplicates().reset_index(drop=True)
ola_data['trip_id'] = ola_data.index
```

```python
ola_data.head()
```

```python
ola_data.info()
```

```python
datetime_dim = ola_data[['pickup_time', 'drop_time']].drop_duplicates().reset_index(drop=True)
datetime_dim['pick_hour'] = datetime_dim['pickup_time'].dt.hour
datetime_dim['pick_day'] = datetime_dim['pickup_time'].dt.day
datetime_dim['pick_month'] = datetime_dim['pickup_time'].dt.month
datetime_dim['pick_year'] = datetime_dim['pickup_time'].dt.year
datetime_dim['pick_weekday'] = datetime_dim['pickup_time'].dt.weekday

datetime_dim['drop_time'] = datetime_dim['drop_time']
datetime_dim['drop_hour'] = datetime_dim['drop_time'].dt.hour
datetime_dim['drop_day'] = datetime_dim['drop_time'].dt.day
datetime_dim['drop_month'] = datetime_dim['drop_time'].dt.month
datetime_dim['drop_year'] = datetime_dim['drop_time'].dt.year
datetime_dim['drop_weekday'] = datetime_dim['drop_time'].dt.weekday

datetime_dim['datetime_id'] = datetime_dim.index

datetime_dim = datetime_dim[['datetime_id', 'pickup_time', 'pick_hour', 'pick_day', 'pick_month', 'pick_year', 'pick_weekday',
                             'drop_time', 'drop_hour', 'drop_day', 'drop_month', 'drop_year', 'drop_weekday']]
#
datetime_dim.head()
```

```python
passenger_count_dim = ola_data[['num_passengers']].reset_index(drop=True)
passenger_count_dim['passenger_count_id'] = passenger_count_dim.index
passenger_count_dim = passenger_count_dim[['passenger_count_id','num_passengers']]

trip_distance_dim = ola_data[['distance']].reset_index(drop=True)
trip_distance_dim['trip_distance_id'] = trip_distance_dim.index
trip_distance_dim = trip_distance_dim[['trip_distance_id','distance']]
```

```python
passenger_count_dim.head()
```

```python
trip_distance_dim.head()
```

```python
rate_code_type = {
    1:"Standard rate",
    2:"JFK",
    3:"Newark",
    4:"Nassau or Westchester",
    5:"Negotiated fare",
    6:"Group ride"
}

rate_code_dim = ola_data[['rate_code']].reset_index(drop=True)
rate_code_dim['rate_code_id'] = rate_code_dim.index
rate_code_dim['rate_code_name'] = rate_code_dim['rate_code'].map(rate_code_type)
rate_code_dim = rate_code_dim[['rate_code_id','rate_code','rate_code_name']]

# rate_code_dim.head()
```

```python
rate_code_dim.head()
```

```python
pickup_location_dim = ola_data[['pickup_loc']].reset_index(drop=True)
pickup_location_dim['pickup_location_id'] = pickup_location_dim.index
pickup_location_dim = pickup_location_dim[['pickup_location_id','pickup_loc']]

dropoff_location_dim = ola_data[['drop_loc']].reset_index(drop=True)
dropoff_location_dim['dropoff_location_id'] = dropoff_location_dim.index
dropoff_location_dim = dropoff_location_dim[['dropoff_location_id','drop_loc']]



```

```python
pickup_location_dim.head()
```

```python
dropoff_location_dim.head()
```

```python
payment_type_name = {
    1:"Credit card",
    2:"Cash",
    3:"No charge",
    4:"Dispute",
    5:"Unknown",
    6:"Voided trip"
}
payment_type_dim = ola_data[['payment_method']].reset_index(drop=True)
payment_type_dim['payment_type_id'] = payment_type_dim.index
payment_type_dim['payment_type_name'] = payment_type_dim['payment_method'].map(payment_type_name)
payment_type_dim = payment_type_dim[['payment_type_id','payment_method','payment_type_name']]
```

```python
payment_type_dim.head()
```

```python
fact_table = ola_data.merge(passenger_count_dim, left_on='trip_id', right_on='passenger_count_id') \
             .merge(trip_distance_dim, left_on='trip_id', right_on='trip_distance_id') \
             .merge(rate_code_dim, left_on='trip_id', right_on='rate_code_id') \
             .merge(pickup_location_dim, left_on='trip_id', right_on='pickup_location_id') \
             .merge(dropoff_location_dim, left_on='trip_id', right_on='dropoff_location_id')\
             .merge(datetime_dim, left_on='trip_id', right_on='datetime_id') \
             .merge(payment_type_dim, left_on='trip_id', right_on='payment_type_id') \
             [['trip_id','vendor_id', 'datetime_id', 'passenger_count_id',
               'trip_distance_id', 'rate_code_id', 'stored_flag', 'pickup_location_id', 'dropoff_location_id',
               'payment_type_id', 'extra_charges', 'mta_tax', 'toll_amount',
               'improvement_charge']]
```
